c) Figure 3.11 shows that entropy and the Gini index are both monotonically increasing on the range [0, 0.5] and they are both monotonically decreasing on the range [0.5, 1]. Is it possible that information gain and the gain in the Gini index favor different attributes? Explain.

- Yes becuase the actual change in information gain is different

```python
def gini(c1, c2):
    return 1 - c1**2 - c2 ** 2

def misclassification_error(probHI):
    return 1 - probHI

def weighted_avg(weights, scores):
    return sum([tup[0]*tup[1] for tup in zip(weights, scores)])
```

### Problem 1
Load the iris sample dataset from sklearn (load iris()) into Python using a
Pandas dataframe. Induce a set of binary Decision Trees with a minimum of
2 instances in the leaves, no splits of subsets below 5, and an maximal tree
depth from 1 to 5 (you can leave other parameters at their defaults). Which
depth values result in the highest Recall? Why? Which value resulted in the
lowest Precision? Why? Which value results in the best F1 score? Explain the
difference between the micro/macro/weighted methods of score calculation.

```python
iris = load_iris()


iris_frame = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])


features = iris_frame[iris_frame.columns[0:-1]]
category = iris_frame.target

X_train, X_test, Y_train, Y_test = model_selection.train_test_split(features,
                                                                    category,
                                                                    test_size=0.2, 
                                                                    random_state=69)

for i in range(5):
    DT = DecisionTreeClassifier(max_depth=i+1, min_samples_split=5, min_samples_leaf=2)
    DT = DT.fit(X_train,
                  Y_train)
    expected = Y_test
    predicted = DT.predict(X_test)
    print('max_depth =', i+1, '\n', "Confusion Matrix:\n",metrics.confusion_matrix(expected,predicted),\
          "\n",metrics.classification_report(expected,predicted))
    
```

- Max depth 3-5 had the highest recall becuase they all yielded the same confusion matrix, the more levels in your decision tree you will get a smaller error rate at each level becuase there is always gain so that means less FP and More TP

- The highest F1 score is again with max depth 3-5 and that's becuase it is "combination" of precision and recall. with every level added to a decision tree there is always info gain

- micro':calculates precision,recall and f1-score by taking raw total of TP FP FN TN.

- 'macro': Does not take into account weight when calculating TP, FP, FN, TN for each label
- 'weighted': Same as macro but takes weighted mean



### Problem 2
Load the Breast Cancer Wisconsin (Diagnostic) sample dataset from the UCI
Machine Learning Repository (The discrete version at: breast-cancer-
wisconsin.data) into Python using a Pandas dataframe. Induce a binary
Decision Tree with a minimum of 2 instances in the leaves, no splits of subsets
below 5, and a maximal tree depth of 2 (use the default Gini criterion). Calculate
the Entropy, Gini, and Misclassification Error of the first split - what is the
Information Gain? What is the feature selected for the first split, and what
value determines the decision boundary?

```python
path = r"https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data"
feature_names=["Sample code number"," Clump Thickness", "Uniformity of Cell Size",\
               "Uniformity of Cell Shape","Marginal Adhesion", "Single Epithelial Cell Size",\
               "Bare Nuclei", "Bland Chromatin", "Normal Nucleoli","Mitoses","Class"]


discrete_df = pd.read_csv(path, na_values=["?"],names=feature_names)
```

```python

imp = SimpleImputer(strategy = "mean")
discrete_df[["Bare Nuclei"]] = imp.fit_transform(discrete_df[["Bare Nuclei"]])

features = discrete_df[discrete_df.columns[0:-2]]
target = discrete_df[["Class"]]
featureTest, featureTrain, targetTest, targetTrain = model_selection.train_test_split(features, 
                                                                                      target,
                                                                                    test_size=0.3,
                                                                                random_state=96)

dt1 = DecisionTreeClassifier(max_depth=1, min_samples_split=5, min_samples_leaf=2)
dt1 = dt1.fit(featureTrain, targetTrain)

expected = targetTest
predicted = dt1.predict(featureTest)
print('max_depth =', 1, '\n', "Confusion Matrix:\n",metrics.confusion_matrix(expected,predicted),\
        "\n",metrics.classification_report(expected,predicted))


dt2 = DecisionTreeClassifier(max_depth=2, min_samples_split=5, min_samples_leaf=2)
dt2 = dt2.fit(featureTrain, targetTrain)

expected = targetTest
predicted = dt2.predict(featureTest)
print('max_depth =', 2, '\n', "Confusion Matrix:\n",metrics.confusion_matrix(expected,predicted),\
        "\n",metrics.classification_report(expected,predicted))
```
